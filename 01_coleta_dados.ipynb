```python
import pandas as pd

caminho_arquivo = "dados/brutos/renda_municipios_ibge.csv"

df_renda_municipios = pd.read_csv(
    caminho_arquivo,
    encoding="utf-8-sig"
)

novos_nomes_colunas = {
    "V": "valor",
    "D1C": "codigo_municipio",
    "D1N": "municipio",
    "D2N": "indicador",
    "D3N": "ano"
}

df_renda_municipios = df_renda_municipios.rename(
    columns=novos_nomes_colunas
)

df_renda_municipios = df_renda_municipios.iloc[1:].copy()

print("Dimensões após limpeza:")
print(df_renda_municipios.shape)

print("\nColunas disponíveis:")
print(df_renda_municipios.columns.tolist())

display(
    df_renda_municipios[
        ["codigo_municipio", "municipio", "valor", "ano"]
    ].head()
)
```

```python
import pandas as pd

df_renda_municipios["valor"] = pd.to_numeric(
    df_renda_municipios["valor"],
    errors="coerce"
)

quantidade_nulos = df_renda_municipios["valor"].isna().sum()

print("Quantidade de valores nulos:", quantidade_nulos)

print("\nTipo da coluna valor:")
print(df_renda_municipios["valor"].dtype)

print("\nEstatísticas do PIB municipal:")
print(df_renda_municipios["valor"].describe())
```

# Insight de Negócio 01 — Concentração Econômica dos Municípios Brasileiros

## Evidência

A mediana do PIB municipal é de aproximadamente R$ 274 milhões, indicando que metade dos municípios brasileiros opera abaixo desse patamar econômico.

Por outro lado, apenas 25% dos municípios superam aproximadamente R$ 756 milhões de PIB.

O contraste entre o menor município analisado (R$ 21 milhões) e o maior (R$ 945 bilhões) revela uma concentração econômica extremamente elevada.

## Interpretação

O potencial de consumo e de concessão de crédito não está distribuído de forma homogênea no território nacional.

Municípios economicamente mais fortes concentram parcela desproporcional da atividade econômica brasileira.

## Aplicação de Negócio

Empresas que distribuem orçamento comercial, equipes de vendas ou campanhas de marketing de forma uniforme entre municípios tendem a superinvestir em regiões de baixa capacidade econômica.

## Decisão Recomendada

Priorizar municípios posicionados nos quartis superiores de atividade econômica para operações de:

- crédito consignado;
- financiamento de veículos;
- energia solar;
- consórcios;
- produtos e serviços de alto ticket.

## Hipótese para Validação nas Próximas Etapas

Municípios com maior PIB apresentarão maior propensão à concessão de crédito e maior potencial de consumo para produtos financeiros.

```python

df_top_20_municipios = (
    df_renda_municipios[
        ["codigo_municipio", "municipio", "valor"]
    ]
    .sort_values(
        by="valor",
        ascending=False
    )
    .head(20)
)

print("Top 20 municípios por PIB:")

display(df_top_20_municipios)

valor_total_top_20 = df_top_20_municipios["valor"].sum()

print(
    f"\nPIB acumulado dos 20 maiores municípios: "
    f"{valor_total_top_20:,.0f}"
)
```

# Insight de Negócio 02 — Polos Econômicos Ocultos Além das Grandes Capitais

## Evidência

Os 20 maiores municípios brasileiros em PIB acumulam aproximadamente R$ 3,17 trilhões em atividade econômica.

Além dos resultados esperados, como São Paulo, Rio de Janeiro, Brasília, Belo Horizonte e Curitiba, o ranking revelou a presença de municípios como:

- Maricá (RJ)
- Saquarema (RJ)
- Osasco (SP)
- Barueri (SP)

## Descoberta

Esses municípios não figuram entre os maiores centros populacionais do país, mas apresentam volume econômico suficiente para competir com capitais estaduais e grandes regiões metropolitanas.

O resultado indica que tamanho populacional e relevância econômica não são necessariamente proporcionais.

## Implicação para Inteligência de Mercado

Estratégias comerciais baseadas exclusivamente em população ou quantidade de domicílios podem ignorar mercados altamente atrativos.

Municípios com elevada concentração econômica tendem a oferecer maior potencial para:

- crédito;
- investimentos;
- seguros;
- consórcios;
- produtos e serviços de alto valor agregado.

## Hipótese para Investigação

Existem municípios brasileiros que concentram atividade econômica significativamente acima do esperado para seu porte populacional.

Caso essa hipótese seja confirmada, a utilização de indicadores econômicos poderá produzir segmentações de mercado mais precisas do que critérios tradicionais baseados apenas em população ou localização geográfica.
